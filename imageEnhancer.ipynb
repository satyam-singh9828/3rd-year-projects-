```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#image load
image = cv2.imread('/content/istockphoto-637135492-612x612.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.subplot(1,2,1)
plt.title('Before')
plt.imshow(image)
#pixel values is increment with 10
brightness = 100

contrast = 2
image2=cv2.addWeighted(image,contrast,np.zeros(image.shape,image.dtype),0,brightness)

cv2.imwrite('contrast.jpg',image2)
plt.subplot(1,2,2)
plt.title('After')
plt.imshow(image2)

plt.show()
```
