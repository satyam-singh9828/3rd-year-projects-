```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def gamma_correction(image, gamma=1.0):

    inv_gamma = 1.0 / gamma
    table = np.array([(i / 255.0) ** inv_gamma * 255 for i in range(256)]).astype("uint8")


    return cv2.LUT(image, table)


img = cv2.imread("/content/food.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


gamma_low = gamma_correction(img, gamma=0.5)   # Brighten
gamma_high = gamma_correction(img, gamma=2.0)  # Darken

# Show results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Gamma = 0.5 (Bright)")
plt.imshow(gamma_low)
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Gamma = 2.0 (Dark)")
plt.imshow(gamma_high)
plt.axis("off")

plt.show()
```
