```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("C:\\Users\\gaura\\OneDrive\\Pictures\\Saved Pictures\\mother.jpg")
img=cv2.resize(img,(400,400))
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist=cv2.calcHist([img],[0],None,[255],[0,255])
plt.plot(hist)
plt.show()
_,thre=cv2.threshold(img,200,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2.imshow("THRESHOLDING",thre)
cv2.imshow("gaurav",img)


cv2.waitKey(0)

```
