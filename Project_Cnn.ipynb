```python
from google.colab import drive
import os

# Mount your Google Drive
drive.mount('/content/drive')

# Set dataset paths (adjust based on your Drive folder)
base_dir = '/content/drive/MyDrive/DIP_dataset'  # Change this to your dataset folder in Drive
train_dir = os.path.join(base_dir, 'train')
val_dir   = os.path.join(base_dir, 'val')
test_dir  = os.path.join(base_dir, 'test')

# Verify folder structure
print("Train folder contents:", os.listdir(train_dir))
print("Validation folder contents:", os.listdir(val_dir))
print("Test folder contents:", os.listdir(test_dir))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import tensorflow.keras.backend as K
import matplotlib.cm as cm

# ===== DIP Enhancement Function =====
def enhance_image(img_path, target_size=(224,224)):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    enhanced = clahe.apply(img)
    enhanced = cv2.resize(enhanced, target_size)
    enhanced = cv2.cvtColor(enhanced, cv2.COLOR_GRAY2RGB)
    enhanced = enhanced / 255.0
    return enhanced

# ===== Prepare file paths and labels =====
def prepare_filepaths_labels(directory):
    filepaths = []
    labels = []
    class_map = {'NORMAL':0, 'PNEUMONIA':1}
    for class_name in ['NORMAL','PNEUMONIA']:
        class_folder = os.path.join(directory, class_name)
        if not os.path.exists(class_folder):
            raise FileNotFoundError(f"Folder not found: {class_folder}")
        for f in os.listdir(class_folder):
            filepaths.append(os.path.join(class_folder,f))
            labels.append(class_map[class_name])
    return filepaths, labels

train_files, train_labels = prepare_filepaths_labels(train_dir)
val_files, val_labels = prepare_filepaths_labels(val_dir)
test_files, test_labels = prepare_filepaths_labels(test_dir)

# ===== Custom generator =====
def generator_with_enhance(filepaths, labels, batch_size=16):
    i = 0
    while True:
        batch_x = []
        batch_y = []
        for b in range(batch_size):
            if i == len(filepaths):
                i = 0
            img = enhance_image(filepaths[i])
            batch_x.append(img)
            batch_y.append(labels[i])
            i += 1
        yield np.array(batch_x), np.array(batch_y)

train_gen = generator_with_enhance(train_files, train_labels)
val_gen = generator_with_enhance(val_files, val_labels)
test_gen = generator_with_enhance(test_files, test_labels)

steps_train = len(train_files)//16
steps_val   = len(val_files)//16
steps_test  = len(test_files)//16

# ===== CNN Model =====
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# ===== Train Model =====
history = model.fit(
    train_gen,
    steps_per_epoch=steps_train,
    validation_data=val_gen,
    validation_steps=steps_val,
    epochs=10
)

# ===== Plot Accuracy & Loss =====
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# ===== Evaluate on Test Set =====
y_pred_probs = []
y_true = test_labels[:steps_test*16]
for _ in range(steps_test):
    x_batch, y_batch = next(test_gen)
    preds = model.predict(x_batch)
    y_pred_probs.extend(preds.ravel())

y_pred = (np.array(y_pred_probs) > 0.5).astype(int)

# Confusion Matrix
cm_matrix = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal','Pneumonia'], yticklabels=['Normal','Pneumonia'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_true, y_pred, target_names=['Normal','Pneumonia']))

# ===== Grad-CAM Visualization =====
def grad_cam(model, img_array, layer_name=None):
    # Automatically pick last Conv2D layer if layer_name not provided
    if layer_name is None:
        for layer in reversed(model.layers):
            if 'conv' in layer.name:
                layer_name = layer.name
                break
    if layer_name is None:
        raise ValueError("No Conv2D layer found in model.")

    # Call model once to ensure layer outputs exist
    _ = model.predict(img_array)

    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:,0]

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = K.mean(grads, axis=(0,1,2))
    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_outputs), axis=-1)
    heatmap = np.maximum(heatmap,0)/np.max(heatmap)
    heatmap = cv2.resize(heatmap.numpy(), (224,224))

    return heatmap

# Sample Grad-CAM visualization
sample_img_path = test_files[0]
img_array = np.expand_dims(enhance_image(sample_img_path), axis=0)
heatmap = grad_cam(model, img_array)

img = cv2.imread(sample_img_path)
img = cv2.cvtColor(cv2.resize(img,(224,224)), cv2.COLOR_BGR2RGB)
heatmap_color = cm.jet(heatmap)[:,:,:3]
superimposed = heatmap_color*0.4 + img/255.0

plt.imshow(superimposed)
plt.title('Grad-CAM Overlay')
plt.axis('off')
plt.show()

```
