```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read image (grayscale)
img = cv2.imread("histoImg.jpg", cv2.IMREAD_GRAYSCALE)

# 1. Global Histogram Equalization
global_eq = cv2.equalizeHist(img)

# Store results
titles = ['Original', 'Global HE']
images = [img, global_eq]

# Plot images and histograms
plt.figure(figsize=(10, 8))

for i in range(len(images)):  # <-- fix here
    # Show image
    plt.subplot(len(images), 2, 2*i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

    # Show histogram
    plt.subplot(len(images), 2, 2*i+2)
    plt.hist(images[i].ravel(), 256, [0, 256], color='black')
    plt.title(titles[i] + " Histogram")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
